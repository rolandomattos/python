##Ejercicio N°1

Escribir un programa que pregunte al usuario por las ventas de los últimos 5 días, y muestre por pantalla una serie con los datos de las ventas por producto: Nro de unidades y precio unitario, indexada por los días, antes y después de aplicarles un descuento del 5%.


### *Respuesta*:



```python
# Importando librerias
import string
import random
import pandas as pd
import numpy as np
import statistics as std
```

```python
# Generando la Data

# marcamos una semilla
random.seed(123)

# Generando el indice con fechas
dates = pd.date_range('1/1/2022', periods=8)

# Generando el String de Productos
productos = ['a', 'b', 'c', 'd', 'e','f','g','h']


# Implementar funciones para obtener datos aleatorios para Cantidad y Precio Unitario

n = 8
def cantidad_data(n):
    cantidad = []
    for i in range(n):
        cantidad.insert(i, random.randrange(10, 100, 2))
    return cantidad


def precio_unit(n):
    preciounit = []
    for i in range(n):
        preciounit.insert(i, random.randrange(1, 10, 2))
    return preciounit


# Generando el DATAFRAME
df = pd.DataFrame({'Producto':productos ,
                   'Cantidad':cantidad_data(n) ,
                   'PrecioUnit': precio_unit(n)},index=dates)
# Le damos nombre al indice
df.index.name="Fecha"

# Insertando la columna Total
df['Total'] = df['Cantidad']*df['PrecioUnit']

# Insertando la columna con el Dscto 5%
df['Dcto5%'] = df['Total']*0.95

# Mostramos todos los datos del DataFrame creado
df
```

```python
# Creamos la interfaz con el usuario para que interactúe sobre las ventas
r = input("Desea ver las ventas de los últimos 5 días (Si/No) : ")
rpt = r.upper()

if(rpt == 'SI'):
    # Definiendo las 5 ultimas ventas
    df_ult_5 = df.tail(5) 
    # Mostramos los datos solicitados
    df_list= df_ult_5[['Producto','Cantidad','PrecioUnit','Total']]
    print("VENTAS DE LOS ÚLTIMOS 5 DÍAS")
    print('=========================================================')
    print("Ventas sin descuento")
    print(df_list)
    print('---------------------------------------------------------')
    print("Ventas con descuento 5%")
    df_list1= df_ult_5[['Producto','Cantidad','PrecioUnit','Dcto5%']]
    print(df_list1)
else:
    print("No hay ventas")
```

Columna 3 =  Si Gasto>Ventas, calcular el sobre gasto.
Obtener el total de las ventas de los últimos 3 meses
Obtener el total de gastos de febrero a abril


### *Respuesta*:

```python
# Generamos el Dataframe
mes = ['Enero','Febrero','Marzo','Abril','Marzo']
ventas = [30500,35600,28300,33900,42500]
gastos = [22000,23450,18100,35700,32450]

df_ej3= pd.DataFrame({'Mes':mes ,
                   'Ventas':ventas,
                   'Gastos': gastos})
# Mostramos el DataFrame
df_ej3
```

```python
# Insertamos las columnas solicitadas

# Insertando la columna 1
df_ej3['%Gastos'] = df_ej3['Gastos']/df_ej3['Ventas']*100

# Insertando la columna 2
df_ej3['Logro'] = ["Meta Superada" if s >=3000 else "Meta No Superada" for s in df_ej3['Ventas']] 

# Insertando la columna 3
df_ej3['Sobre_Gasto'] = np.where(df_ej3['Gastos']>df_ej3['Ventas'], df_ej3['Gastos']-df_ej3['Ventas'] , None)

df_rspt= df_ej3.drop(['Status','Orden'], axis=1)

df_rspt

# Mostrando el DataFrame

#df_ej3.groupby(by=['Orden','Mes']).sum()


```

```python
# Creamos una agrupación por mes

# Añadimos un orden para los meses
df_rspt['Orden'] = [1,2,3,4,3]

# Agrupamos por mes
df_rspt1= df_rspt.groupby(by=['Orden','Mes']).sum()

# Mostramos el Dataframe agrupado por mes
df_rspt1
```

```python
# Obtenemos los ultimos 3 meses
df_rspt2= df_rspt1.tail(3)

# Mostramos la data obtenida
df_rspt2
```

```python
# Obtenemos el total de las Ventas de los ultimos 3 meses
ventas_ult = sum(df_rspt2['Ventas'])
ventas_ult
```

```python
# Obtenemos el total de los gastos de febrero a abril
gastos_ult = sum(df_rspt2['Gastos'])
gastos_ult
```
